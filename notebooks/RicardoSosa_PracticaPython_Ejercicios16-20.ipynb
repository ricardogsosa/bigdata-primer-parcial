# Ejercicio 16: DataFrame de estudiantes

### Enunciado:
Crear un DataFrame con información de estudiantes incluyendo nombre, carrera, edad, promedio y ciudad. Luego mostrar información general, calcular promedios y ordenar registros.

### Lógica:
Se crea un diccionario con la información de los estudiantes y posteriormente se convierte en un DataFrame utilizando Pandas. Después se utilizan funciones como head(), info(), mean() y sort_values() para mostrar información, calcular promedios y ordenar los datos.

```python
import pandas as pd

datos = {
    "nombre": ["Carlos", "Andrea", "Luis", "María", "José",
               "Fernanda", "Kevin", "Sofía", "Miguel", "Valeria"],

    "carrera": ["Ingeniería", "Medicina", "Derecho", "Arquitectura",
                 "Ingeniería", "Psicología", "Contaduría",
                 "Marketing", "Diseño", "Ingeniería"],

    "edad": [20, 22, 21, 23, 19, 24, 22, 20, 25, 21],

    "promedio": [85, 91, 76, 88, 95, 79, 82, 90, 73, 87],

    "ciudad": ["Tegucigalpa", "San Pedro Sula", "La Ceiba",
                "Choluteca", "Comayagua", "Danlí",
                "El Progreso", "Juticalpa",
                "Puerto Cortés", "Tela"]
}

df = pd.DataFrame(datos)

print("Primeras 5 filas:")
print(df.head())

print("\nInformación general:")
print(df.info())

print("\nPromedio general:")
print(df["promedio"].mean())

print("\nEstudiantes con promedio mayor o igual a 80:")
print(df[df["promedio"] >= 80])

print("\nEstudiantes ordenados por promedio:")
print(df.sort_values(by="promedio", ascending=False))
```

### Explicación:
En este ejercicio se utilizó Pandas para crear y manipular un DataFrame con información de estudiantes. Se aplicaron funciones básicas para mostrar registros, obtener información general, calcular promedios y ordenar datos. Además, se utilizaron filtros para identificar estudiantes con calificaciones mayores o iguales a 80, permitiendo realizar un pequeño análisis académico de forma organizada.

# Ejercicio 17: Análisis de ventas con pandas

### Enunciado:
Crear un DataFrame de ventas y realizar cálculos de ingresos, agrupaciones y análisis básicos sobre los productos y vendedores.

### Lógica:
Se construye un DataFrame con registros de ventas. Luego se crea una nueva columna llamada TotalVenta multiplicando la cantidad por el precio unitario. Finalmente se utilizan agrupaciones y ordenamientos para analizar vendedores, productos y categorías.

```python
import pandas as pd

ventas = {
    "fecha": ["2026-01-01", "2026-01-02", "2026-01-03", "2026-01-04",
              "2026-01-05", "2026-01-06", "2026-01-07", "2026-01-08",
              "2026-01-09", "2026-01-10", "2026-01-11", "2026-01-12",
              "2026-01-13", "2026-01-14", "2026-01-15"],

    "vendedor": ["Ana", "Luis", "Pedro", "Ana", "Luis",
                  "Pedro", "Ana", "Luis", "Pedro", "Ana",
                  "Luis", "Pedro", "Ana", "Luis", "Pedro"],

    "producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Laptop",
                  "Mouse", "Tablet", "Monitor", "Laptop", "Teclado",
                  "Tablet", "Mouse", "Laptop", "Monitor", "Tablet"],

    "categoria": ["Tecnología", "Accesorios", "Accesorios", "Tecnología",
                   "Tecnología", "Accesorios", "Tecnología",
                   "Tecnología", "Tecnología", "Accesorios",
                   "Tecnología", "Accesorios", "Tecnología",
                   "Tecnología", "Tecnología"],

    "cantidad": [2, 5, 3, 1, 2, 4, 1, 2, 1, 3, 2, 5, 1, 2, 3],

    "precio_unitario": [25000, 500, 1200, 7000, 25000,
                         500, 15000, 7000, 25000, 1200,
                         15000, 500, 25000, 7000, 15000]
}

df = pd.DataFrame(ventas)

df["TotalVenta"] = df["cantidad"] * df["precio_unitario"]

print("Total vendido por vendedor:")
print(df.groupby("vendedor")["TotalVenta"].sum())

print("\nProducto más vendido:")
print(df.groupby("producto")["cantidad"].sum())

print("\nCategoría con mayor ingreso:")
print(df.groupby("categoria")["TotalVenta"].sum())

print("\nVentas ordenadas:")
print(df.sort_values(by="TotalVenta", ascending=False))
```

### Explicación:
En este ejercicio se desarrolló un análisis básico de ventas utilizando un DataFrame. Se creó una nueva columna para calcular el total de cada venta multiplicando la cantidad por el precio unitario. También se utilizaron agrupaciones para identificar vendedores, productos y categorías con mayores ingresos, facilitando la interpretación de los datos comerciales.

# Ejercicio 18: Limpieza de datos con pandas

### Enunciado:
Crear un DataFrame con datos incompletos y aplicar procesos básicos de limpieza de datos.

### Lógica:
Se crea un DataFrame con valores vacíos y nulos. Después se identifican los valores faltantes utilizando isnull(), se reemplazan ciudades vacías y se eliminan registros que no contienen correo electrónico.

```python
import pandas as pd

clientes = {
    "nombre": ["Mario", "Lucía", "Carlos", "Andrea", "Pedro"],

    "edad": [25, 30, None, 28, 35],

    "correo": ["mario@gmail.com", None,
                "carlos@gmail.com", "andrea@gmail.com", None],

    "ciudad": ["Tegucigalpa", "", "La Ceiba", "", "Choloma"],

    "telefono": ["9999-1111", "8888-2222",
                  None, "7777-3333", "6666-4444"]
}

df = pd.DataFrame(clientes)

print("Valores nulos:")
print(df.isnull())

print("\nCantidad de nulos por columna:")
print(df.isnull().sum())

df["ciudad"] = df["ciudad"].replace("", "No especificada")

df = df.dropna(subset=["correo"])

print("\nDataFrame limpio:")
print(df)
```

### Explicación:
En este ejercicio se trabajó con limpieza de datos utilizando funciones de Pandas para detectar valores nulos y datos incompletos. Se reemplazaron valores vacíos en la columna de ciudad y se eliminaron registros que no contenían correo electrónico. Esto permitió obtener un DataFrame más limpio y adecuado para futuros análisis.

# Ejercicio 19: Análisis de préstamos

### Enunciado:
Crear un DataFrame con solicitudes de préstamos y realizar análisis básicos sobre los montos y estados de aprobación.

### Lógica:
Se almacena la información de préstamos en un DataFrame y luego se utilizan funciones estadísticas y filtros para calcular promedios, contar solicitudes y mostrar únicamente ciertos registros según condiciones específicas.

```python
import pandas as pd

prestamos = {
    "cliente": ["Mario", "Lucía", "Carlos", "Andrea",
                "Pedro", "Fernanda", "Luis", "Valeria"],

    "monto_solicitado": [25000, 60000, 45000, 80000,
                         30000, 52000, 70000, 40000],

    "plazo": [12, 24, 18, 36, 12, 24, 36, 18],

    "ingreso_mensual": [15000, 30000, 18000, 40000,
                        16000, 28000, 35000, 20000],

    "estado": ["Aprobado", "Pendiente", "Rechazado",
               "Aprobado", "Pendiente", "Aprobado",
               "Rechazado", "Aprobado"]
}

df = pd.DataFrame(prestamos)

print("Monto promedio solicitado:")
print(df["monto_solicitado"].mean())

print("\nSolicitudes por estado:")
print(df["estado"].value_counts())

print("\nPréstamos aprobados:")
print(df[df["estado"] == "Aprobado"])

print("\nClientes con préstamos mayores a 50,000:")
print(df[df["monto_solicitado"] > 50000])

print("\nPromedio por estado:")
print(df.groupby("estado")["monto_solicitado"].mean())
```

### Explicación:
En este ejercicio se realizó un análisis de solicitudes de préstamos mediante filtros y funciones estadísticas de Pandas. Se calcularon promedios, se contaron solicitudes según su estado y se identificaron clientes con montos elevados. Además, se utilizaron agrupaciones para analizar el promedio solicitado por cada estado de préstamo.

# Ejercicio 20: Mini análisis de datos tipo empresa

### Enunciado:
Crear un DataFrame de empleados y realizar análisis relacionados con salarios, experiencia y departamentos.

### Lógica:
Se crea un DataFrame con información de empleados. Posteriormente se utilizan funciones estadísticas, filtros y agrupaciones para analizar salarios y experiencia laboral. También se crea una nueva columna para clasificar el nivel de experiencia.

```python
import pandas as pd

empleados = {
    "nombre": ["Carlos", "Andrea", "Luis", "María",
                "Pedro", "Fernanda", "Kevin", "Sofía"],

    "departamento": ["IT", "RRHH", "Ventas", "IT",
                     "Finanzas", "Ventas", "IT", "RRHH"],

    "cargo": ["Programador", "Analista", "Supervisor",
              "Desarrollador", "Contador",
              "Vendedor", "Administrador", "Asistente"],

    "salario": [35000, 28000, 32000, 40000,
                30000, 26000, 45000, 25000],

    "experiencia": [2, 5, 7, 10, 4, 6, 8, 1],

    "ciudad": ["Tegucigalpa", "San Pedro Sula",
                "La Ceiba", "Comayagua",
                "Danlí", "Choluteca",
                "Tela", "Juticalpa"]
}

df = pd.DataFrame(empleados)

print("Primeros registros:")
print(df.head())

print("\nEstadísticas de salarios:")
print(df["salario"].describe())

print("\nSalario promedio por departamento:")
print(df.groupby("departamento")["salario"].mean())

print("\nEmpleado con salario más alto:")
print(df[df["salario"] == df["salario"].max()])

print("\nEmpleados con más de 5 años de experiencia:")
print(df[df["experiencia"] > 5])

print("\nEmpleados ordenados por salario:")
print(df.sort_values(by="salario", ascending=False))

def nivel(exp):
    if exp <= 2:
        return "Junior"
    elif exp <= 5:
        return "Intermedio"
    else:
        return "Senior"

df["NivelExperiencia"] = df["experiencia"].apply(nivel)

print("\nDataFrame actualizado:")
print(df)
```

### Explicación:
En este ejercicio se creó un DataFrame con información de empleados para realizar análisis relacionados con salarios y experiencia laboral. Se aplicaron estadísticas descriptivas, filtros y ordenamientos para identificar empleados destacados y promedios por departamento. También se creó una nueva columna para clasificar automáticamente el nivel de experiencia de cada empleado.
